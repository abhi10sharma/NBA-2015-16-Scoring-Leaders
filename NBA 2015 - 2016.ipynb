```python
import pandas as pd

nba_data = pd.read_excel('.xlsx')
nba_data = nba_data.drop(['Image', 'Team'], axis = 1)
nba_data.tail()
```

```python
players_to_plot = ['Steph Curry', 'James Harden', 'Kevin Durant', 'Lebron James', 'Kawhi Leonard']

start_date = '27/10/2015'
end_date = '13/04/2016'
```

```python
import matplotlib.pyplot as plt
import datetime

def get_index(date):
    first_date = datetime.datetime(2015, 10, 27)
    
    date_list = [int(x) for x in date.split('/')]
    date_obj = datetime.datetime(date_list[-1], date_list[1], date_list[0])

    return (date_obj - first_date).days

start_index = get_index(start_date)
end_index = get_index(end_date)

dates = list(nba_data)[1:]

x_values = dates[start_index: end_index + 1]
plt.figure(figsize=(6,4), dpi = 150)

for players in players_to_plot:
    player_df = nba_data[nba_data['Player'] == players]
    y_values = [int(player_df[col]) for col in x_values]
    plt.plot(x_values, y_values, label = players)
    
plt.title('NBA 2015-2016 Scoring Leaders', fontdict = {'fontname': 'Comic Sans MS', 'fontsize':25, 'color': '#800606'}) 
plt.xlabel('Dates', fontsize = 20)
plt.ylabel('Points', fontsize = 20)
plt.legend() 
plt.savefig('nba 2015-16 scoring leaders.png', bbox_inches='tight')
plt.show()    
```
